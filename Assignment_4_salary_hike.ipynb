Model Building

```python
from scipy import stats
import numpy as np
import statsmodels.formula.api as smf
```

```python
model1=smf.ols("Salary~Experience",data=df).fit()
```

```python
print(model1.params,'\n''\n',model1.tvalues,'\n''\n',model1.pvalues)
```

```python
model2=smf.ols("Salary~np.log(Experience)",data=df).fit() # use log
print(model2.params,'\n''\n',model2.tvalues,'\n''\n', model2.pvalues)
```

```python
(model2.rsquared,model2.rsquared_adj)#rsquared values
```

```python
model2.summary()
```

```python
model3=smf.ols("Salary~np.square(Experience)",data=df).fit() #using square
```

```python
print(model3.params,'\n''\n',model3.tvalues,model3.pvalues)
```

```python
(model3.rsquared,model3.rsquared_adj)
```

```python
model3.summary()
```

```python
model4=smf.ols("Salary~np.sqrt(Experience)",data=df).fit() #using square
```

```python
print(model4.params,'\n''\n',model4.tvalues,model4.pvalues)
```

```python
(model4.rsquared,model4.rsquared_adj)
```

```python
model4.summary()
```
