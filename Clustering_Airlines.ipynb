```python
## Hierarchial Clustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
h_data=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
h_data
```

```python
from sklearn.preprocessing import MinMaxScaler
Scale=MinMaxScaler()
h_data_norm=Scale.fit_transform(h_data)
h_data_norm
```

```python
h_data_norm = pd.DataFrame(h_data_norm)
h_data_norm
```

```python
from scipy.cluster.hierarchy import linkage
link=linkage(h_data_norm,method='average',metric='euclidean')
link
```

```python
from scipy.cluster.hierarchy import dendrogram
dend=dendrogram(link)
```

```python
from sklearn.cluster import AgglomerativeClustering
Agg_clust=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='average')
```

```python
Agg_clust.fit(h_data_norm)
```

```python
Agg_clust.get_params
```

```python
Agg_clust.labels_
```

```python
h_labels=Agg_clust.labels_
```

```python
h_data['clusters']=h_labels
```

```python
h_data
```

```python
h_data.iloc[:,1:].groupby(h_data['clusters']).mean()
```

```python
h_data['clusters'].value_counts()
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(h_data_norm,h_labels)
score
```

# Performing K Means Clustering

```python

k_data=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
k_data
```

```python
##Standardise data
scale=MinMaxScaler()
k_data_norm=scale.fit_transform(k_data)
k_data_norm=pd.DataFrame(k_data_norm,columns=k_data.columns)
k_data_norm
```

```python
##Plot elbow curve for K value
WCSS=[]
from sklearn.cluster import KMeans
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model=model.fit(k_data_norm)
    WCSS.append(model.inertia_)
plt.plot(range(1,10),WCSS)
```

```python
WCSS
```

```python
##From the above curve, we can use k value as 5
```

```python
##build KMeans cluster model with no of cluster as 5
k_model=KMeans(n_clusters=5)
k_model=k_model.fit(k_data_norm)
k_model
```

```python
k_clust_labels=k_model.labels_
```

```python
k_data['Clust_labels']=k_clust_labels
k_data
```

```python
k_data['Clust_labels'].value_counts()
```

```python
k_data.plot(x='Bonus_miles',y='Bonus_trans',kind='scatter',s=50,c=k_clust_labels,cmap=plt.cm.BrBG_r)
```

```python
##Calculating Silhoutte SCore to check accuracy of labels
from sklearn.metrics import silhouette_score
k_accuracy=silhouette_score(k_data_norm,k_clust_labels)
k_accuracy
```

# Performing DBScan Clustering

```python
db_data=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
db_data
```

```python
##Standardise the given data:
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
db_data_norm=scale.fit_transform(db_data.iloc[:,1:])
db_data_norm=pd.DataFrame(db_data_norm)
db_data_norm
```

```python
##Find Epsilon Radius using Nearest Neighbours
from sklearn.neighbors import NearestNeighbors
neigh=NearestNeighbors(n_neighbors=22)
neigh=neigh.fit(db_data_norm)
distances,indices=neigh.kneighbors(db_data_norm)
```

```python
distances
```

```python
distances=np.sort(distances,axis=0)
```

```python
plt.plot(distances[:,1])
```

```python
##We can consider epsilon value of 0.2
```

```python
from sklearn.cluster import DBSCAN
db_model=DBSCAN(eps=0.2,min_samples=12)
db_model=db_model.fit(db_data_norm)
```

```python
db_model.components_
```

```python
db_model_labels=pd.DataFrame(db_model.labels_,columns=['DBScan_Labels'])
db_model_labels
```

```python
db_data['DBSCan Labels']=db_model_labels
db_data=db_data.drop(columns='ID#',axis=1)
db_data
```

```python

db_data.plot(x='Bonus_miles',y='Bonus_trans',cmap=plt.cm.gist_rainbow,kind='scatter',s=80,c=db_data['DBSCan Labels'],figsize=(20,20))
```

```python
##Calculate Silhooutte score for clusters
from sklearn.metrics import silhouette_score
db_accuracy=silhouette_score(db_data_norm,db_model_labels)
db_accuracy
```

```python
##Accuracy Table:
index=['KMeans Score','DBScan Score']
columns=['Silhoutte Score']
labels=[k_accuracy,db_accuracy]
pd.DataFrame(labels,index=index,columns=columns)
```
